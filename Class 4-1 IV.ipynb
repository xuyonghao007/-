```python
whole_data=pd.read_csv("whole_data_female.csv")
whole_data
```

```python
df=whole_data[['Female_mng_per','TotalSalary','Female','Age','Female_board_per', 'tobinq', 'asset', 'roa', 'debt_asset', 'IsDuality',"Female_monitor_per",'Indirect_per']]
df["TotalSalary"] = (df["TotalSalary"]+1).apply(np.log1p)
df["asset"] = (df["asset"]+1).apply(np.log1p)
df=df.dropna()
```

```python
x1 = df[['Female','Age', 'tobinq', 'asset', 'roa', 'debt_asset', 'IsDuality','Indirect_per']]
x1 = sm.add_constant(x1) 
y = df['TotalSalary']
```

```python
result = sm.OLS(y, x1, missing="drop").fit()
print(result.summary())
```

```python
IV= df[['Female_mng_per','Indirect_per','Female_board_per','Age', 'tobinq', 'asset', 'roa', 'debt_asset', 'IsDuality']]
IV = sm.add_constant(IV) 
```

```python

from statsmodels.sandbox.regression.gmm import IV2SLS
import statsmodels.api as sm
import statsmodels.formula.api as smf
import linearmodels.iv.model as lm

```

```python
result_IV = IV2SLS(y,
                  x1, #所有原始自变量
                  IV).fit() #工具变量，还包含除去内生的自变量以外的全部自变量           
```

```python
print(result_IV.summary())
```

```python
exog_var = df[['Age', 'tobinq', 'asset', 'roa', 'debt_asset', 'IsDuality','Indirect_per']]

iv_var=df[['Female_mng_per','Female_monitor_per']] #,'Female_monitor_per'
mlr2 = lm.IV2SLS(y, 
                 exog_var, 
                 df["Female"], 
                 iv_var).fit()
#mlr2.summary.as_csv()
```
